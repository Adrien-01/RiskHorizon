**Description des tables**

- **application_{train|test}.csv**

Il s'agit du tableau principal, divisé en deux fichiers : « Train » (avec TARGET) et « Test » (sans TARGET).
Données statiques pour toutes les applications. Chaque ligne correspond à un prêt de notre échantillon de données.

- **bureau.csv**

Tous les crédits antérieurs du client accordés par d'autres établissements financiers et signalés à l'agence d'évaluation du crédit (pour les clients ayant un prêt dans notre échantillon).
Pour chaque prêt de notre échantillon, le nombre de lignes correspond au nombre de crédits que le client avait enregistrés auprès de l'agence d'évaluation du crédit avant la date de la demande.

- **bureau_balance.csv**

Soldes mensuels des crédits antérieurs enregistrés auprès de l'agence d'évaluation du crédit.
Ce tableau comporte une ligne pour chaque mois d'historique de tous les crédits antérieurs déclarés à l'agence d'évaluation du crédit ; en d'autres termes, le tableau comporte (nombre de prêts de l'échantillon × nombre de crédits antérieurs correspondants × nombre de mois pour lesquels nous disposons d'un historique observable concernant ces crédits) lignes.

- **POS_CASH_balance.csv**

Aperçus mensuels des soldes des anciens crédits à la consommation et des prêts en espèces que le demandeur avait contractés auprès de Home Credit.
Ce tableau comporte une ligne pour chaque mois de l'historique de chaque crédit antérieur chez Home Credit (crédits à la consommation et prêts en espèces) lié aux prêts de notre échantillon – c'est-à-dire que le tableau comporte (nombre de prêts de l'échantillon × nombre de crédits antérieurs correspondants × nombre de mois pour lesquels nous disposons d'un historique observable pour ces crédits antérieurs) lignes.

- **credit_card_balance.csv**

Aperçus mensuels des soldes des anciennes cartes de crédit que le demandeur a souscrites auprès de Home Credit.
Ce tableau comporte une ligne pour chaque mois d'historique de chaque crédit antérieur chez Home Credit (crédits à la consommation et prêts en espèces) lié aux prêts de notre échantillon – c'est-à-dire que le tableau comporte (nombre de prêts de l'échantillon × nombre de cartes de crédit antérieures concernées × nombre de mois pour lesquels nous disposons d'un historique observable pour la carte de crédit antérieure) lignes.

- **previous_application.csv**

 Toutes les demandes antérieures de prêt Home Credit des clients ayant contracté un prêt dans notre échantillon.
Il y a une ligne pour chaque demande antérieure relative aux prêts figurant dans notre échantillon de données.

- **installments_payments.csv**

Historique des remboursements des crédits Home Credit déjà versés et liés aux prêts de notre échantillon.
Il y a :
a) une ligne pour chaque paiement effectué, ainsi que
b) une ligne pour chaque paiement manqué.
Une ligne correspond à un paiement d'une mensualité OU à une mensualité correspondant à un paiement d'un crédit Home Credit antérieur lié aux prêts de notre échantillon.

- **HomeCredit_columns_description.csv**

Ce fichier contient les descriptions des colonnes des différents fichiers de données.

# Master Dataset - Table application_train

Pour les besoins du projet, on allons conserver uniquement 20 colonnes pour créer un **Master Dataset** à partir de la table **application_train**

###### 1. La Variable Cible & Identifiants (Le Cœur du Projet)

* **`SK_ID_CURR`** : L'identifiant unique de la demande de crédit (servira pour les jointures et pour servir de clé primaire).
* **`TARGET`** : La variable binaire la plus importante (**0** = crédit remboursé sans problème, **1** = client en difficulté de paiement). C'est ce que nous allons chercher à analyser.

###### 2. Les Variables Financières (Notre terrain de jeu)

* **`NAME_CONTRACT_TYPE`** : Déterminer s'il s'agit d'un prêt en espèces ou d'un crédit renouvelable
* **`AMT_INCOME_TOTAL`** : Le revenu annuel global du client.
* **`AMT_CREDIT`** : Le montant total du crédit accordé par la banque.
* **`AMT_ANNUITY`** : Le montant annuel (!) (de la mensualité) que le client doit rembourser.
* **`AMT_GOODS_PRICE`** : Le prix du bien que le client achète avec le crédit (ex: le prix de la maison ou de la voiture).

###### 3. Le Profil Démographique (Pour créer la dimension `dim_clients`)

* **`CODE_GENDER`** : Sexe du client.
* **`DAYS_BIRTH`** : L'âge du client (exprimé en jours négatifs, parfait pour un petit nettoyage en Python : `df['AGE'] = df['DAYS_BIRTH'] / -365`).
* **`CNT_CHILDREN`** : Le nombre d'enfants à charge.
* **`NAME_EDUCATION_TYPE`** : Le niveau d'études (secondaire, académique, etc.).
* **`NAME_FAMILY_STATUS`** : Situation familiale (marié, célibataire, divorcé).
* **`NAME_INCOME_TYPE`** : La source de revenu (salarié, indépendant, retraité, demandeur d'emploi).

###### 4. Le Profil Professionnel & Patrimoine (Pour la dimension `dim_biens`)

* **`DAYS_EMPLOYED`** : L'ancienneté dans l'emploi actuel (en jours négatifs également).
* **`FLAG_OWN_CAR`** / **`FLAG_OWN_REALTY`** : Est-ce que le client possède déjà sa voiture ou son logement ? (Y/N).
* **`ORGANIZATION_TYPE`** : Le secteur d'activité de l'employeur (Banque, Construction, Éducation...). Ultra-puissant pour simuler un impact de crise sectorielle lors du *Stress Test*.

###### 5. Les Scores de Risque Externes (servira pour le ML)

* **`EXT_SOURCE_1`**, **`EXT_SOURCE_2`**, **`EXT_SOURCE_3`** : Trois scores standardisés provenant d'organismes de crédit externes. Ce sont des variables continues entre 0 et 1. Ils devront faire une moyenne ou un score combiné en Python pour évaluer la fiabilité globale du profil.

#### **Suppression des lignes vides AMT_ANNUITY - 12 lignes**

La mensualité (AMT_ANNUITY) est l'élément central pour calculer le taux d'endettement et le reste à vivre. Si elle est manquante sur une demande de crédit active, le dossier est inexploitable d'un point de vue bancaire.

````python
df_credit = df_credit.dropna(subset=["AMT_ANNUITY"])
df_credit.shape
````

#### **Que faire des lignes vides de AMT_GOODS_PRICE (~300 valeurs nulles) ?**

Avant de supprimer ces 300 lignes, il faut regarder à quel type de crédit elles correspondent. En effet, dans le dataset Home Credit, il y a deux grands types de prêts (colonne NAME_CONTRACT_TYPE) :


*   Cash loans (Prêts de trésorerie / Prêts personnels) : La banque vous verse 5 000 € sur votre compte courant et vous en faites ce que vous voulez. Dans ce cas, il n'y a pas de "bien" acheté à l'origine, donc AMT_GOODS_PRICE peut légitimement être nul ou vide.
*   Revolving loans (Crédits renouvelables) : Une réserve d'argent disponible. Là aussi, aucun prix de bien n'est rattaché lors de la demande initiale.

La bonne pratique bancaire consiste à dire : "Si le prix du bien n'est pas spécifié, on considère que le prix du bien est égal au montant du crédit accordé (AMT_CREDIT)".

````python
df_credit["AMT_GOODS_PRICE"] = df_credit["AMT_GOODS_PRICE"].fillna(df_credit["AMT_CREDIT"])
df_credit.shape
````

#### **Pourquoi laisser les valeurs nulles dans `EXT_SOURCE_1`, `2` et `3` ?**

Cela évite de tomber dans un piège statistique très classique qui risque de fausser complètement les analyses ainsi que les prédictions de l'IA.

Les variables `EXT_SOURCE` sont des scores de crédit normalisés provenant d'organismes externes. Ils sont compris strictement **entre 0 et 1** (par exemple : `0.15` pour un profil très risqué, `0.85` pour un profil excellent).

* **Un score de 0 a une signification métier très lourde :** Si un client obtient un score de `0`, cela signifie qu'il est considéré par les organismes externes comme le pire profil possible, un fraudeur ou quelqu'un en faillite totale.
* **Le piège :** Si un client n'a tout simplement pas de score externe disponible (par exemple parce qu'il n'a jamais contracté de crédit auparavant, ce qui arrive souvent), lui attribuer un `0` va artificiellement le faire passer pour un profil ultra-dangereux aux yeux de la banque. Cela va complètement perturber le modèle de Machine Learning lors de son entraînement.

#### **Harmonisation des formats des colonnes de "string"** `

L'harmonisation consiste à supprimer les espaces masqués invisibles avec la méthode `.str.strip()` et standardiser le casse avec la méthode `.upper()`.
On remplace également les caractères spéciaux par un "_"

````python
def harmoniser_texte(df, liste_colonnes):
    for col in liste_colonnes:
        if col in df.columns:
            df[col] = (
                df[col].astype(str).str.strip().str.upper().str.replace(" ", "_")
            )
    return df


# Utilisation
colonnes_textes = ["NAME_CONTRACT_TYPE", "ORGANIZATION_TYPE", "CODE_GENDER","NAME_INCOME_TYPE","NAME_EDUCATION_TYPE",	"NAME_FAMILY_STATUS"]
df_credit = harmoniser_texte(df_credit, colonnes_textes)
````

#### **Traitement des valeurs vides**

**Valeurs vides dans `DAYS_CREDIT_ENDDATE`**

La colonne `DAYS_CREDIT_ENDDATE` indique le nombre de jours restants avant la fin théorique de ces crédits extérieurs (exprimé en jours négatifs ou positifs par rapport à la date de la demande actuelle).

Avoir des valeurs nulles (vides) ici est tout à fait normal d'un point de vue métier, mais **il ne faut surtout pas les supprimer ni les remplacer par 0**. Voici l'enquête métier et la méthode propre pour les traiter dans **RiskHorizon**.

---

🔍 L'Analyse Métier : Pourquoi ces cases sont-elles vides ?

Dans le Bureau du Crédit, il y a principalement deux types de lignes (colonne `CREDIT_ACTIVE`) :

1. **Closed (Crédits clôturés) :** Le crédit est remboursé et terminé.
2. **Active (Crédits en cours) :** Le client est toujours en train de rembourser.

Si tu croises `DAYS_CREDIT_ENDDATE` avec le statut du crédit, tu vas vite t'apercevoir du scénario suivant :

* **Cas des Crédits Renouvelables (Revolving) ou Découverts :** Lorsqu'un client a une ligne de crédit renouvelable active (une réserve d'argent) ou une autorisation de découvert dans une autre banque, **il n'y a pas de date de fin contractuelle**. Le crédit reste ouvert tant que le client ne résilie pas le contrat. La date de fin est donc légitimement **vide (nulle)**.
* **Cas des vieux crédits mal archivés :** Parfois, pour de vieux crédits clôturés, l'information de la date de fin n'a tout simplement pas été transmise au Bureau.

Si tu remplaces les valeurs manquantes par `0`, tu indiques artificiellement à ton modèle SQL et à ton IA que **le crédit se termine aujourd'hui même**.
Pour un crédit renouvelable en cours, cela fausserait complètement le calcul de l'exposition au risque de ton client à l'instant T.

Si l'équipe veut utiliser cette colonne pour savoir si le client a des engagements à très long terme, elle peut créer une variable binaire lors de l'agrégation dans BigQuery :

```sql
-- Requête d'agrégation dans BigQuery pour le Jour 6
SELECT
    SK_ID_CURR,
    COUNT(SK_ID_BUREAU) AS nb_credits_exterieurs,
    -- On compte combien de crédits n'ont pas de date de fin (souvent des crédits renouvelables actifs)
    COUNTIF(DAYS_CREDIT_ENDDATE IS NULL) AS nb_credits_sans_fin_contractuelle,
    -- Pour la moyenne des jours restants, BigQuery va ignorer les NULL tout seul
    AVG(DAYS_CREDIT_ENDDATE) AS jours_moyens_restants_fin_credit
FROM `votre_projet.bureau`
GROUP BY SK_ID_CURR;

Cela donne une excellente indication à notre IA sur les engagements flexibles du client auprès de la concurrence. »*

**Valeurs vides dans `DAYS_ENDDATE_FACT`**

La colonne `DAYS_ENDDATE_FACT` indique le nombre de jours entre la fermeture effective du crédit extérieur et la demande de crédit actuelle auprès de ta banque. C'est une donnée exprimée en jours négatifs (par exemple, `-180` signifie que le crédit s'est terminé il y a 6 mois).

Les cases vides correspondent à 100 % aux crédits encore ouverts (Active). Il est impossible d'avoir une date de fermeture pour un crédit en cours.

Mettre une valeur ou supprimer ces lignes serait une grosse erreur de manipulation de données.

---

**Possibles utilisations de cette colonne :**
Créer une métrique de "Fraîcheur des Incidents" (Feature Engineering)

Une idée brillante pour le projet est de chercher si le client a fermé un crédit très récemment. Un crédit fermé à la hâte juste avant de demander un nouveau prêt peut être un signal à surveiller.

Voici la requête SQL à intégrer dans **BigQuery** pour nettoyer et agréger intelligemment :

```sql
SELECT
    SK_ID_CURR,
    -- 1. Nombre de crédits actuellement actifs (où DAYS_CREDIT_FACT est logiquement NULL)
    COUNTIF(CREDIT_ACTIVE = 'Active') AS nb_credits_externes_actifs,
    
    -- 2. Nombre de crédits clôturés
    COUNTIF(CREDIT_ACTIVE = 'Closed') AS nb_credits_externes_clotures,
    
    -- 3. Depuis combien de jours le dernier crédit a-t-il été fermé ?
    -- (MAX sur des jours négatifs donne la date la plus proche de 0, donc la plus récente)
    MAX(DAYS_ENDDATE_FACT) AS jours_depuis_derniere_cloture
FROM `votre_projet.bureau`
GROUP BY SK_ID_CURR;

```

#### **Gestion des aberrations numériques**

**Nettoyer les montants négatifs de `AMT_CREDIT_SUM` et `AMT_CREDIT_SUM_DEBT`.**

Un montant de crédit ne peut être négatif, on force les montants négatifs à 0.
On utilise la méthode `.clip`, elle permet de plafonner et plancher les valeurs d'une colonne ou d'un tableau entre deux bornes (un minimum et un maximum).

````python
df_bureau['AMT_CREDIT_SUM'] = df_bureau['AMT_CREDIT_SUM'].clip(lower=0)
df_bureau['AMT_CREDIT_SUM_DEBT'] = df_bureau['AMT_CREDIT_SUM_DEBT'].clip(lower=0)
````

````python
df_bureau.shape
````

**Traiter le cas des dettes supérieures au crédit accordé**

Parfois, en raison des agios ou des retards de paiement, la dette dépasse légèrement le crédit de base. Mais si la dette est 10 ou 100 fois supérieure, c'est une anomalie.
On la plafonne au montant du crédit initial

````python
mask_aberrant = df_bureau['AMT_CREDIT_SUM_DEBT'] > (df_bureau['AMT_CREDIT_SUM'] * 2)
df_bureau.loc[mask_aberrant, 'AMT_CREDIT_SUM_DEBT'] = df_bureau.loc[mask_aberrant, 'AMT_CREDIT_SUM']
````

````python
df_bureau.shape
````

#### **Standardisation des textes (Harmonisation)**

On lisse les colonnes `CREDIT_ACTIVE` (statut du crédit) et `CREDIT_TYPE`, c'est-à-dire que l'on nettoie le casse et les espaces.

````python
df_bureau['CREDIT_ACTIVE'] = df_bureau['CREDIT_ACTIVE'].astype(str).str.strip().str.upper()
df_bureau['CREDIT_TYPE'] = df_bureau['CREDIT_TYPE'].astype(str).str.strip().str.upper().str.replace(" ", "_")
````

**Forcer le casse et supprimer les espaces**

L'harmonisation consiste à supprimer les espaces masqués invisibles avec la méthode .str.strip() et standardiser le casse avec la méthode .upper(). On remplace également les caractères spéciaux par un "_"

````python
def harmoniser_texte(df, liste_colonnes):
    for col in liste_colonnes:
        if col in df.columns:
            df[col] = (
                df[col].astype(str).str.strip().str.upper().str.replace(" ", "_")
            )
    return df

colonnes_textes = ["NAME_CONTRACT_TYPE", "WEEKDAY_APPR_PROCESS_START", "NAME_SELLER_INDUSTRY","NAME_YIELD_GROUP", "PRODUCT_COMBINATION"]
df_previous_application = harmoniser_texte(df_previous_application, colonnes_textes)
````
