```python
from google.colab import drive
import pandas as pd
import gdown

# Montez votre Google Drive
drive.mount('/content/drive')
```

```python

# ID du fichier extrait du lien Google Drive
file_id = '1_Ym2w09qJ01lY6rWAWvbffOnKcUi9_Fc'

# On reconstruit l'URL au format attendu par gdown
url = f"https://drive.google.com/uc?id={file_id}"

# Chemin de destination pour le fichier téléchargé
output_file = '/tmp/Taux_endettement.csv'

print("Téléchargement du fichier depuis Google Drive...")
# gdown télécharge le fichier dans ton dossier local et affiche une barre de progression
gdown.download(url, output_file, quiet=False)

print("Chargement et filtrage du dataset en cours...")
# 4. On lit le fichier localement avec Pandas (beaucoup plus rapide et stable !)
df = pd.read_csv(output_file)

print("Terminé ! Ton DataFrame 'df_credit' est prêt.")
```

```python
df.head()
```

```python

import numpy as np

# Créer une copie
df_copy = df.copy()

# Calculer le taux d'endettement
df_copy['tx_endettement'] = round(((df_copy['AMT_ANNUITY']*12) / df_copy['AMT_INCOME_TOTAL']) * 100, 2)

df_copy['tx_endettement']
```

```python
# Compte les lignes vides'
nb_manquants = df_copy['tx_endettement'].isnull().sum()
print(f"Nombre de valeurs manquantes : {nb_manquants}")
```

```python
# Mettre à jour le DataFrame
df = df_copy.copy()
```
